```python
import csv
import json
import requests
from datetime import datetime
import pickle


token = "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"
headers = {'Authorization': "Token " + token}
rate_limit_url = "https://api.github.com/rate_limit"
req = requests.get(rate_limit_url, headers=headers).json()
print(req)

# print(str(req['message']=='Not Found'))
```

```python
ac_PR = "https://api.github.com/repos/apache/camel/pulls?state=closed&per_page=10&page=1"
PR_req = requests.get(ac_PR, headers=headers).json()
print(PR_req)
with open("prs.txt", 'a') as out:
    out.truncate(0)
    out.write(str(PR_req))
    
```

```python
#get user-login, created-at, merged-at, commits-url, comments-url
prlist = []
for pull in PR_req:
    if(pull["merged_at"]!= None):
        pr = {}
        pr['user_login'] = pull['user']['login']
        pr['created_at'] = pull['created_at']
        pr['merged_at'] = pull['merged_at']
        pr['commits_url'] = pull['commits_url']
        pr['comments_url'] = pull['comments_url']
        prlist.append(pr)
print(len(prlist))
print(prlist)


    
```

```python
#get commits, comments
for i in range(0,len(prlist)):
    commits_req = requests.get(prlist[i]['commits_url'], headers=headers).json()
    comments_req = requests.get(prlist[i]['comments_url'], headers=headers).json()
    print()
    print(len(comments_req))
    prlist[i]['commits'] = len(commits_req)
    prlist[i]['comments'] = len(comments_req)
    print(prlist[i]['comments'])
    
```

```python
#get full-name
for i in range(0,len(prlist)):
    fullname_req = requests.get("https://api.github.com/users/"+prlist[i]['user_login'],headers=headers).json()
    prlist[i]['full_name'] = fullname_req['name']
    print(prlist[i]['full_name'])
```

```python
#get the earlist comments. Since no comments found from camel, create on list to compare.

for i in range(0,len(prlist)):
    if(prlist[i]['comments']==0):
        prlist[i]['first_comment_at'] = 'None'
    else:
        comments_req = requests.get(prlist[i]['comments_url'], headers=headers).json()
        first_comment_at = ''
        for j in range(0,len(comments_req)):
            if(first_comment_at == ''):
                first_comment_at = comments_req[j]['created_at']
            else:
                f = "%Y-%m-%dT%H:%M:%SZ"
                prev = datetime.strptime(first_comment_at, f)
                nex = datetime.strptime(comments_req[j]['created_at'], f)
                if(nex<prev):
                    first_comment_at = comments_req[j]['created_at']
        prlist[i]['first_comment_at'] = first_comment_at
        
        
print(prlist)

                   
```

```python
#write prlist to a file
with open('extracted_prlist.txt', 'wb') as f:
    pickle.dump(prlist, f)
#read prlist from a file
with open('extracted_prlist.txt', 'rb') as f:
    my_list = pickle.load(f)
    print(my_list)
```
